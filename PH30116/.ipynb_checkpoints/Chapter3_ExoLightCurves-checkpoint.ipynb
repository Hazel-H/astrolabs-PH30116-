Dividing by the median to get a normalised lightcurve 

```python
#THis works....
for lcfile in glob.glob('Data/Object%slc/kplr*.fits' %(mykepler)):
    tmp = fits.open(lcfile)
    tmptime = (tmp[1].data['TIME'])
    tmpflux = (tmp[1].data['PDCSAP_FLUX'])
    medianflux = np.nanmedian(tmp[1].data['PDCSAP_FLUX'])
    tmperror = (tmp[1].data['PDCSAP_FLUX_ERR'])
    
    #pylab.ylim(0.997,1.006)
    #pylab.figure(figsize=(15,5))
    #pylab.plot(tmptime, tmpflux/medianflux, c='k')
        
    data = (np.array([tmptime, tmpflux/medianflux]).transpose())
    
    pylab.plot(data[:,0], data[:, 1])
    
    #print((tmptime))
    #print(tmpflux)
```

```python


#THis works....
for lcfile in glob.glob('Data/Object%slc/kplr*.fits' %(mykepler)):
    tmp = fits.open(lcfile)
    tmptime = (tmp[1].data['TIME'])
    tmpflux = (tmp[1].data['PDCSAP_FLUX'])
    medianflux = np.nanmedian(tmp[1].data['PDCSAP_FLUX'])
    tmperror = (tmp[1].data['PDCSAP_FLUX_ERR'])
    
    pylab.ylim(0.997,1.006)
    #pylab.figure(figsize=(8,5))
    #pylab.plot(tmptime, tmpflux/medianflux, c='k')

    pylab.plot(tmptime, tmpflux/medianflux, ls='None', marker='.', c='grey', label='Data')    

```

Practising applying filters to one of the datasets 

```python
from scipy.signal import savgol_filter

time = lc[1].data['TIME']
flux = lc[1].data['PDCSAP_FLUX']
normal_flux = flux/(np.nanmedian(lc[1].data['PDCSAP_FLUX']))
#normal_flux = normal_flux[np.logical_not(np.isnan(normal_flux))]

data = np.array([time, normal_flux]).transpose()
clean_data = data[~np.isnan(data).any(axis=1)]

pylab.figure(figsize=(12,8))
pylab.plot(clean_data[:,0], clean_data[:,1], ls='None', marker='o')

interp_savgol = savgol_filter(normal_flux, window_length=1001, polyorder=3)
pylab.plot(time, interp_savgol, ls='--', c='r',label='Savitzky Golay w=11,o=3')
```

```python
from scipy.signal import medfilt

testfil = medfilt(normal_flux, kernel_size=201) ##kernel_size is the size of the window over which the median is calculated 

pylab.plot(time, normal_flux, c='k', ls='None', marker='o', label='Data')
pylab.plot(time, testfil, ls='--', c='r', label='Median Filter')
```
